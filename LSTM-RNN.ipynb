```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adadelta
from sklearn.model_selection import train_test_split
import pandas as pd

# Data
X = np.arange(1, 50)
Y = X * 15

print("Sequence:", X, Y, sep="\n")

# Reshape for LSTM
X = X.reshape(-1, 1, 1)
Y = Y.reshape(-1, 1)

# Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, shuffle=False)

# Model function
def get_model():
    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(1, 1)))
    model.add(Dense(1))
    model.compile(optimizer=Adadelta(learning_rate=0.03), loss='mse')
    return model

# Set seed
tf.random.set_seed(0)

# Build model
model = get_model()
print(model.summary())

# Train
model.fit(X_train, Y_train, epochs=2000, validation_split=0.2, batch_size=10)

# Save model
tf.keras.models.save_model(model, "./models/seq_pred_model.h5")

# Load model
model = tf.keras.models.load_model("./models/seq_pred_model.h5")

# Prediction
Y_pred = model.predict(X_test, verbose=0)
Y_diff = Y_test - Y_pred

# Result table
result = pd.DataFrame(
    np.hstack([Y_test, Y_pred, Y_diff]),
    columns=["Actual", "Predicted", "Difference"]
)

print(result)
```
